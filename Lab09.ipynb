```python
def merge(lis1,lis2):
    if len(lis1) == 0:
        return lis2
    if len(lis2) == 0:
        return lis1
    if lis1[0] < lis2[0]:
        return [lis1[0]] + merge(lis1[1:],lis2)
    else:
        return [lis2[0]] + merge(lis1,lis2[1:])
def final_merge(g_list):
    if len(g_list) == 1:
        return g_list[0]
    else:
        g_list.insert(0,merge(g_list[0],g_list[1]))
        g_list = [g_list[0]] + g_list[3:]
        return final_merge(g_list)

m = int(input("Number of rows: "))
n = int(input("Number of columns: "))
matrix = []
for i in range (m):
    matrix.append(list(map(int,input().split())))
k = int(input())
if k < m * n:
    for x in matrix:
        if len(x) != n:
            ValueError
    else:
        print(final_merge(matrix)[k-1])
else:
    ValueError
```

```python
def find_min(number):
    min = number[0]
    for n in number:
        if min > n:
            min = n
    return min

def largest_min_index(number):
    min_index = []
    for i in range (len(number)):
            if number[i] == find_min(number):
                min_index.append(i)
    return min_index[-1]

def SelectionSort(origin):
    number = origin.copy()
    for i in range (len(number)):
        index_min = i
        min_index = number[i]
        for j in range (i+1,len(number)):
            if min_index > number[j]:
                min_index = number[j]
                index_min = j
        number[i],number[index_min] = number[index_min],number[i]     
    return number   
 
def swap(number):
    if number == SelectionSort(number):
        number[-2],number[-1] = number[-1],number[-2]
        return number
    if number[0] != find_min(number):
        number[0],number[largest_min_index(number)] = number[largest_min_index(number)],number[0]
        while number[0] == 0:
            number.pop(0)
        return number
    else:
        return [number[0]] + swap(number[1:])
    
a = str(int(input()))
number1 = []
for d in a:
    number1.append(int(d))
for x in swap(number1):
    print(x,end="")
```

```python
def upper(a,b,i=0):
    if i == len(a):
        return ""
    else:
        ch = a[i]
        if ch.upper() in b:
            ch = ch.upper()
        return ch + upper(a,b,i+1)
def answer(n):
    if n == 0:
        return ""
    a = input()
    b = input()
    c = ""
    for x in upper(a,b,i=0):
        if 65 <= ord(x) <= 90:
            c = c + x
    if b in c:
        ans = "YES" 
    else:
        ans = "NO" 
    return ans + "\n" + answer(n-1)
        
Q = int(input())
print(answer(Q))

```

```python
def sort(a):
    for i in range (len(a)):
        min_index = i
        min = a[i]
        for j in range (i+1,len(a)):
            if min > a[j]:
                min = a[j]
                min_index = j
        a[i],a[min_index] = a[min_index],a[i]
    return a
def next_num(a):
    nxt_index = None
    for i in range (1,len(a)):
        if a[i] > a[0]:
            nxt = a[i]
            nxt_index = i
            break
    if nxt_index is None:
        return a
    for i in range (1,len(a)):
        if a[i] > a[0] and a[i] < nxt:
            nxt = a[i]
            nxt_index = i
    a[0],a[nxt_index] = a[nxt_index],a[0]
    return a

def Hoan_vi(lis,n):
    if n == 1:
        lis.reverse()
        return lis
    if lis[n-1] > lis[n-2]:
        lis = lis[:(n-2)] + next_num(lis[(n-2):])
        lis = lis[:(n-1)] + sort(lis[(n-1):])
        return lis
    else:
        return Hoan_vi(lis,n-1)
    
n = int(input())
new_lis = list(map(int,input().split()))
if len(new_lis) == n:
    for x in Hoan_vi(new_lis,n):
        print(x,end=" ")
else:
    ValueError

            
```

```python
def Permutation(lis):
    for i in range (len(lis)-2,-1,-1):
        if lis[i] < lis[i+1]:
            for j in range (len(lis)-1,i,-1):
                if lis[i] < lis[j]:
                    lis[i],lis[j] = lis[j],lis[i]
                    break
            lis = lis[0:i] + [lis[i]] + sorted(lis[(i+1):])
            return lis
    lis.reverse()
    return lis

n = int(input())
new_lis = list(map(int,input().split()))
if len(new_lis) == n:
    for x in Permutation(new_lis):
        print(x,end=" ")
else:
    ValueError
```

```python
def check(arr):
    L = 0
    R = 0
    U = 0
    D = 0
    for x in arr:
        if x == "L":
            L += 1
        elif x == "R":
            R += 1
        elif x == "U":
            U += 1
        else:
            D += 1
    return L == R and U == D

def is_path_crossing(moves): 
    n = len(moves)
    for length in range (2,n+1,2):
        for j in range (0,n-length+1):
            if check(moves[j:j+length]):
                return True
    return False

moves = input()
print(is_path_crossing(moves))

    
    
```

```python
def min(x,y):
    if x < y:
        min_xy = x
    else:
        min_xy = y
    return min_xy

def max(x,y):
    if x < y:
        max_xy = y
    else:
        max_xy = x
    return max_xy

def Median(nums):
    a,b,c,d,e = nums
    f = min(min(a,b),min(c,d))
    g = max(max(a,b),max(c,d))
    h = max(min(a,b),min(c,d))
    i = min(max(a,b),max(c,d))
    
    if e < f:
        return min(h,i)
    else:
        if e > g:
            return max(h,i)
        else:
            if e > max(h,i):
                return max(h,i)
            elif e < min(h,i):
                return min(h,i)
            else:
                return e

nums = list(map(int,input().split()))
print(Median(nums))
```

```python
def findRadius(houses, heaters):
    min_distance = []
    for x in houses:
        min = abs(x - heaters[0])
        for y in heaters:
            if min > abs(x-y):
                min = abs(x-y)
        min_distance.append(min)

    max = 0
    for x in min_distance:
        if max < x:
            max = x
    r = max
    return r

houses = list(map(int,input().split()))
heaters =list(map(int,input().split()))
print(findRadius(houses,heaters))
```
