```python
sequence = list(map(int,input().split()))
n = int(input())
def binary_search(sequence, n):
    left, right = 0, len(sequence) - 1
    while left <= right:
        mid = (left + right) // 2
        if sequence[mid] == n:
            return mid
        elif sequence[mid] > n:
            right = mid - 1
        else:
            left = mid + 1
    return -1

print(binary_search(sequence,n))
```

```python
arr = list(map(int,input().split()))
x = int(input())
def count_occurence(arr,x):
    left,right = 0,len(arr) - 1 
    count = 0
    while left < right:
        if arr[left] == x: 
            count += 1
        if arr[right] == x:
            count += 1
        left += 1
        right -= 1
    return count
print(count_occurence(arr,x))

```

```python
arr = list(map(int,input().split()))
def bubble_sort(arr):
    for i in range (len(arr) - 1):
        swapped = False
        for j in range (len(arr) - i - 1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
                swapped = True
        if not swapped:
            break
    return arr
for x in bubble_sort(arr):
    print(x,end=" ")
  
```

```python
arr = list(map(int,input().split()))
def selection_sort(arr):
    for i in range (len(arr)):
        min = i
        for j in range (i+1,len(arr)):
            if arr[min] > arr[j]:
                min = j
        arr[i],arr[min] = arr[min],arr[i]
                
    return arr
for x in selection_sort(arr):
    print(x,end=" ")
```

```python
from heapq import heappush, heappop

def minInterval(intervals, queries):
    # Sort intervals by left endpoint
    intervals.sort(key=lambda x: x[0])

    # Sort queries, keep original index
    sorted_queries = sorted([(q, i) for i, q in enumerate(queries)])

    ans = [-1] * len(queries)
    heap = []  # (length, r)
    j = 0      # pointer for intervals

    for q, qi in sorted_queries:
        # Add all intervals with l <= q
        while j < len(intervals) and intervals[j][0] <= q:
            l, r = intervals[j]
            heappush(heap, (r - l + 1, r))
            j += 1

        # Remove intervals that can't cover q (r < q)
        while heap and heap[0][1] < q:
            heappop(heap)

        # The top is the smallest length interval that still can cover q
        if heap:
            ans[qi] = heap[0][0]
        else:
            ans[qi] = -1

    return ans
```
