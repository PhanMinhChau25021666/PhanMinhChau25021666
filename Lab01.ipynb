W2A15

```python
def star_number(n):
    if n <= 0:
        raise ValueError("n la so nguyen duong")
    return 6*n*(n-1)+1

def main():
    try:
        n = int(input("Nhap so nguyen duong n: "))
        
        if n <= 0:
            print("Vui long nhap so nguyen duong!")
            return
        
        result = star_number(n)
        print(f"So sao thu{n} la: {result}")

        print("\nMot so sao dau tien: ")
        for i in range(1, min(n+1, 11)):
            print(f"S({i}) = {star_number(i)}")

    except ValueError:
        print("Vui long nhap mot so nguyen hop le!")

if __name__ == "__main__":
    main()
    
```
