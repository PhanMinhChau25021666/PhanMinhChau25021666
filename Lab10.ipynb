```python
def checking(a):
    a1 = 0
    while a > 0: 
        a1 += (a%10)**2
        a = a//10
    return a1
next = set()
def HappyNumbers(a):
    if a == 1:
        return "YES"
    if a in next:
        return "NO"
    else:
        next.add(a)
    a = checking(a)
    return HappyNumbers(a)

try:
    a = int(input())
    if a <= 0:
        raise ValueError
    print(HappyNumbers(a))
except ValueError:
    print("Error: Invalid value")



```

```python
N = int(input())
lis1 = list(map(int,input().split()))

def check(n,lis):
    if n <= 0:
        raise ValueError
    if len(lis) != n:
        raise ValueError   
    if len(set(lis)) < n:
        raise ValueError

try:
    check(N,lis1)
except:
    print("Mang khong hop le")
else:
    sum = 0 
    for x in lis1:
        sum += x
    print(sum)
```

```python
def checking(S):
    lis = []
    for x in S:
        lis.append(x)
    for i in range (len(lis)-1):
        if lis[i] == "*" or lis[i] == "+" or lis[i] == "?":
            if lis[i+1] == "*" or lis[i+1] == "+" or lis[i+1] == "?":
                return False
    return True
try:
    T = int(input())
    for i in range (T):
        S = input()
        print(checking(S))
except:
    print(False)
```
