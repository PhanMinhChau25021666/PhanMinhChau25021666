```python
n = int(input("Nhap so nguyen duong n >= 2: "))
if n < 2:
    ValueError
else:
    factors = list()
    for num in range (2, n+1):
        if n % num == 0:
            factors.append(num)
    
    prime_factors = list()
    for factor in factors:
        for i in range (2,factor):
            if factor % i == 0:
                break
        else:
            prime_factors.append(factor)

    max_prime = prime_factors[-1]
    prime_factors.remove(prime_factors[-1])
    for prime in prime_factors:
        if prime > max_prime:
            max_prime = prime
        prime_factors.remove(prime)
    print(max_prime)


```
