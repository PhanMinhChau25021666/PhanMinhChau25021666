```python
c = input("Nhap vao ki tu c: ")
if len(c) == 1:
    if "A" <= c <= "Z" and 65 <= ord(c) <= 90:
        print(c.lower())
        print(chr(ord(c)+32))
    elif "a" <= c <= "z" and 97 <= ord(c) <= 122:
        print(c.upper())
        print(chr(ord(c)-32))
    else:
        ValueError
else:
    IndexError
```

```python
c = input("Nhap vao ki tu c: ")
if len(c) == 1:
    if 65 <= ord(c) <= 90 or 65 <= ord(c) - 32 <= 90:
        print(f"{c} la ki tu alphabet")
    else:
        print(f"{c} khong la ki tu alphabet")
else:
    IndexError
```

```python
A = input("Nhap vao chu cai hoa A: ")
if len(A) == 1:
    if 65 < ord(A) <= 90:
        print(chr(ord(A) + 31))
    elif ord(A) == 65:
        print(chr(ord(A) + 32))
    else:
        ValueError
else:
    IndexError
```

```python
a, b, c = map(float, input("Nhap vao 3 so: ").split())
if a > 0 and b > 0 and c > 0: 
    if a < b + c and b < c + a and c < a + b:
        p = (a + b + c)/2
        S = (p * (p-a) * (p-b) * (p-c))**(1/2)
        print(f"{S:.1f}")
    else:
        print("Khong phai 3 canh cua tam giac")
else:
    ValueError
          
```

```python
Ten = input("Ten chu ho: ")
truoc = int(input("Chi so dien ke thang truoc: "))
nay = int(input("Chi so dien ke thang nay: "))
tong = nay - truoc
tien = 0
if 0 <= tong <= 50:
    tien = tong * 1984
elif 51 <= tong <= 100:
    tien = 50 * 1984 + (tong - 50) * 2050
elif 101 <= tong <= 200:
    tien = 50 * 1984 + 50 * 2050 + (tong - 100) * 2380
elif 201 <= tong <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (tong - 200) * 2998
elif 301 <= tong <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (tong - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (tong - 400) * 3460
print(f"Ho va ten: {Ten}")
print(f"Tien phai tra la: {tien * 1.08:.0f}")


```
