```python
class Cylinder:
    def __init__(self,r,h):
        self.r = r
        self.h = h
    def surface_area(self):
        return 2 * 3.14 * self.r * (self.h + self.r)
    def Volume(self):
        return 3.14 * (self.r)**2 * self.h
r,h = map(float,input().split())
cy = Cylinder(r,h)
print(f"{cy.surface_area():.2f}")
print(f"{cy.Volume():.2f}")
```

```python
class Date:
    def __init__ (self,dd,mm,yyyy):
        self.d = dd
        self.m = mm
        self.y = yyyy

    def next_day (self):
        if self.m % 2 == 1:
            if 1 <= self.m <= 7:
                if 1 > self.d or self.d > 31:
                    return "INVALID"
                elif self.d == 31:
                    return f"1/{self.m + 1}/{self.y}"
                else:
                    return f"{self.d + 1}/{self.m}/{self.y}"

            elif 9 <= self.m <= 11:
                if 1 > self.d or self.d > 30:
                    return "INVALID"
                elif self.d == 30:
                    return f"1/{self.m + 1}/{self.y}"
                else:
                    return f"{self.d + 1}/{self.m}/{self.y}"                    
            else:
                return "INVALID"
        else:
            if self.m == 2:
                if self.y % 4 == 0: 
                    if 1 > self.d or self.d > 29:
                        return "INVALID"
                    elif self.d == 29:
                        return f"1/3/{self.y}"  
                    else:
                        return f"{self.d + 1}/{self.m}/{self.y}"                   
                else:
                    if 1 > self.d or self.d > 28:
                        return "INVALID"
                    elif self.d == 28:
                        return f"1/3/{self.y}"  
                    else:
                        return f"{self.d + 1}/{self.m}/{self.y}"
                                            
            elif 4 <= self.m <= 6:
                if 1 > self.d or self.d > 30:
                    return "INVALID"
                elif self.d == 30:
                    return f"1/{self.m + 1}/{self.y}"
                else:
                    return f"{self.d + 1}/{self.m}/{self.y}"  
            elif 8 <= self.m <= 12:
                if 1 > self.d or self.d > 31:
                    return("INVALID")
                elif self.d == 31:
                    if self.m == 12:
                        return f"1/1/{self.y + 1}"
                    else:
                        return f"1/{self.m + 1}/{self.y}"
                else:
                    return f"{self.d + 1}/{self.m}/{self.y}"  
            else:
                return "INVALID"
    
            
d,m,y = map(int,input().split("/"))
date = Date(d,m,y)
print(date.next_day())
```

```python
class ShoppingCart:
    def __init__ (self):
        self.item = []
    def add (self,name,price):
        self.item.append({"name":name,"price":price})
    def remove (self,name):
        for item in self.item:
            if name in item:
                self.item.remove(item)
    def empty (self):
        return len(self.item) == 0
    def total_price (self):
        sum = 0
        for item in self.item:
            sum += item["price"]
    def display (self):
        for item in self.item:
            print(item)
    def remove (self):
        self.item.clear()

cart = ShoppingCart()
while True:
    print("Menu")
    print("2. Add")
    print("3. Remove")
    print("4. Empty")
    print("5. Total price")
    print("6. Display")
    print("7. Remove")

    choice = int(input())
    if choice == 2:
        name = input()
        price = float(input())
        cart.add(name,price)
    elif choice == 3:
        name = input()
        cart.remove(name)
    elif choice == 4:
        cart.empty()
    elif choice == 5:
        cart.total_price()
    elif choice == 6:
        cart.display()
    elif choice == 7:
        cart.remove()
    else:
        ValueError
    
    exit = input("Có muốn thoát khỏi chương trình không?").strip()
    if exit.lower() == "yes":
        break
```

```python
class Rectangle:
    def __init__ (self,width,height):
        self.w = width
        self.h = height
    def area (self):
        return self.w * self.h
    def perimeter (self):
        return 2 * (self.w + self.h)
    def scale (self,k):
        self.w *= k
        self.h *= k
w,h,k = map(float,input().split())
rec = Rectangle(w,h)
rec.scale(k)
print(f"{rec.area():.2f}",f"{rec.perimeter():.2f}")
        
```

```python
import math

class Fraction:
    def __init__(self,num,den):
        if den == 0:
            ValueError
        if den < 0:
            num = -num
            den = -den
        common = math.gcd(num,den)
        self.num = num // common
        self.den = den // common

    def add (self,self1):
        add_num = self.num*self1.den + self1.num*self.den
        add_den = self.den*self1.den
        return Fraction(add_num,add_den)

    def sub (self,self1):
        sub_num = self.num*self1.den - self1.num*self.den
        sub_den = self.den*self1.den
        return Fraction(sub_num,sub_den)
    
    def mul (self,self1):
        mul_num = self.num*self1.num
        mul_den = self.den*self1.den
        return Fraction(mul_num,mul_den)
    
    def div (self,self1):
        div_num = self.num*self1.den
        div_den = self.den*self1.num
        return Fraction(div_num,div_den)
    
    def __str__ (self):
        return f"{self.num}/{self.den}"
    
a, b , op, c, d = map(str, input().split())
a,b,c,d = map(int,[a,b,c,d])
if op == "+":
    ab = Fraction(a,b)
    cd = Fraction(c,d)
    print(ab.add(cd))
if op == "-":
    ab = Fraction(a,b)
    cd = Fraction(c,d)
    print(ab.sub(cd))
if op == "*":
    ab = Fraction(a,b)
    cd = Fraction(c,d)
    print(ab.mul(cd))
if op == "/":
    ab = Fraction(a,b)
    cd = Fraction(c,d)
    print(ab.div(cd))

        


```

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.o = owner
        self.b = balance
    def deposit (self, x):
        self.b += x
    def withdraw (self, x):
        self.b -= x
    def __str__(self):
        return f"{self.o}: {self.b:.2f}"
o, b = map(str,input().split())
b = float(b)
n = int(input())
acc = BankAccount(o,b)
for i in range (n):
    dw, a = map(str,input().split())
    a = float(a)
    if dw == "DEPOSIT":
        acc.deposit(a)
    elif dw == "WITHDRAW":
        acc.withdraw(a)
print(acc)
```

```python
class Student:
    def __init__(self,name):
        self.n = name
        self.t = []

    def add_score(self,subject,score):
        self.t.append({subject:score})
    def avg(self):
        if len(self.t) > 0:
            total = 0
            for i in self.t:
                for x in i.values():
                    total += x
            self.aver = total/(len(self.t))
            return self.aver
    def rank(self):
        if self.aver >= 8:
            return "Excellent"
        elif 6.5 <= self.aver < 8:
            return "Good"
        elif 5 <= self.aver < 6.5:
            return "Average"
        else:
            return "Poor"
    def __str__ (self):
        return f"{self.n} {self.avg():.2f} {self.rank()}"

name = input()
n = int(input())
stu = Student(name)
for i in range (n):
    sub,scr = map(str,input().split())
    scr = float(scr)
    stu.add_score(sub,scr)

stu.avg()
stu.rank()
print(stu)

```

```python
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.books = []
    def add(self,book):
        self.books.append(book)
    def count_by_author(self,name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count
    def find_by_year(self,y):
        count = 0
        for boo in self.books:
            if boo.year == y:
                count += 1
        return count
    
n = int(input())
libra = Library()
for i in range (n-1):
    add_book = input().split()
    add_book.pop(0)
    t, a, y = map(str,add_book[0].split(";"))
    book = Book(t,a,y)
    libra.add(book)

count_method, n_or_y = map(str,input().split())
if count_method == "COUNT":
    print(libra.count_by_author(n_or_y))
elif count_method == "COUNTYEAR":
    print(libra.find_by_year(n_or_y))
```

```python
import math

class Surface:
    def __init__(self):
        self.PQR = []
    def add_point(self,x,y,z):
        self.PQR.append([x,y,z])
    def n_vector (self):
        PQ = [self.PQR[1][0] - self.PQR[0][0] , self.PQR[1][1] - self.PQR[0][1], self.PQR[1][2] - self.PQR[0][2]]
        PR = [self.PQR[2][0] - self.PQR[0][0] , self.PQR[2][1] - self.PQR[0][1], self.PQR[2][2] - self.PQR[0][2]]
        self.n = [PQ[1]*PR[2]-PQ[2]*PR[1],PQ[2]*PR[0]-PQ[0]*PR[2],PQ[0]*PR[1]-PQ[1]*PR[0]]

class Degree:
    def __init__(self):
        pass
    def angle(self,sur1,sur2):
        cos_phi = (sur1.n[0]*sur2.n[0] + sur1.n[1]*sur2.n[1] + sur1.n[2]*sur2.n[2])/(((sur1.n[0]**2 + sur1.n[1]**2 + sur1.n[2]**2)**(1/2))*((sur2.n[0]**2 + sur2.n[1]**2 + sur2.n[2]**2)**(1/2)))
        return f"{math.degrees(math.acos(cos_phi)):.2f}"
    
point = []
for i in range (4):
    point.append(list(map(float,input().split())))

sur1 = Surface()
sur2 = Surface()

for i in range (0,3):
    sur1.add_point(point[i][0],point[i][1],point[i][2])
for i in range (1,4):
    sur2.add_point(point[i][0],point[i][1],point[i][2])

sur1.n_vector()
sur2.n_vector()
print(Degree().angle(sur1,sur2))

```

```python
class Triangle:
    def __init__(self,initEdge1,initEdge2,initEdge3):
        self.e1 = int(initEdge1)
        self.e2 = int(initEdge2)
        self.e3 = int(initEdge3)
    def getArea(self):
        a,b,c = self.e1,self.e2,self.e3
        if a > 0 and b > 0 and c > 0 and a + b > c and b + c > a and c + a > b: 
            p = (a + b + c)/2
            return (p*(p-a)*(p-b)*(p-c))**(1/2)
        else:
            return -1
        
a,b,c = map(float,input().split())
area = Triangle(a,b,c).getArea()
if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
