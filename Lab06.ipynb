```python
x = input().strip('{}').split(", ")
y = input().strip('{}').split(", ")
z = x + y
if "" in z:
    z.remove("")
z.sort()
ans = {}
for i in z:
    key,value = i.split(": ")
    key = key.strip("'\'")
    value = int(value)
    if key not in ans:
        ans[key] = value
    else:
        ans[key] += value

print(ans)
```

```python
import ast

d1 = ast.literal_eval(input())   # nhập dict 1
d2 = ast.literal_eval(input())   # nhập dict 2

result = {}

# cộng gộp dict 1
for k, v in d1.items():
    result[k] = v

# cộng gộp dict 2
for k, v in d2.items():
    if k in result:
        result[k] += v
    else:
        result[k] = v

# in theo thứ tự tăng dần của key
for k in sorted(result):
    print(f"{k}: {result[k]}")
```

W6A10

```python
Z = list(map(int,input().split()))
k = int(input())
match = []
ans = []
for i in range (len(Z)):
    for j in range (i+1,len(Z)):
        if Z[i] + Z[j] == k:
            match.append([Z[i],Z[j]])
match1 = []
for n in match:
    if n not in match1:
        match1.append(n)
match1.sort()
for i in range (len(match1)):
    x = tuple(match1[i])
    ans.append(x)
print(ans)
```

```python
Z = list(map(int,input().split()))
Z.sort()
counting = []
for i in range (len(Z)):
    count = 0
    for j in range (i+1,len(Z)):
        if Z[i] == Z[j]:
            count += 1
    counting.append(count)
max = 0
for i in range (len(Z)):
    if max < counting[i]:
        max = counting[i]
for i in range (len(counting)):
    if max == counting[i]:
        print(Z[i])
        break

```
