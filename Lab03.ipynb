```python
Watts = float(input("Nhập vào số kWh điện tiêu thụ: "))
if 0 <= Watts <= 50:
    So_tien = Watts*1500
elif 51 <= Watts <= 100:
    So_tien = 50*1500 + (Watts - 50)*2000
else:
    So_tien = 50*1500 + 50*2000 + (Watts-100)*3000
print(f"Tiền điện: {So_tien}")
```

```python
so_thuc = float(input("Nhap vao 1 so thuc: "))
phan_nguyen = int(so_thuc)
thap_phan = so_thuc - phan_nguyen
if thap_phan == 0:
    up = so_thuc
    down = so_thuc
    depend = so_thuc

elif thap_phan > 0: 
    up = phan_nguyen + 1
    down = phan_nguyen
    if (thap_phan*10) <5:
        depend = down
    else:
        depend = up

else:
    up = phan_nguyen
    down = phan_nguyen - 1
    if (abs(thap_phan)*10) <5:
        depend = up
    else:
        depend = down

print(up, down, depend)
```
